```python
import cv2
import os
import numpy as np
from pathlib import Path
from tqdm import tqdm

def canny_sketch(img, low=35, high=110):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    clahe = cv2.createCLAHE(clipLimit=1.5, tileGridSize=(8, 8))
    equalized = clahe.apply(gray)
    filtered = cv2.bilateralFilter(equalized, d=7, sigmaColor=65, sigmaSpace=65)
    edges = cv2.Canny(filtered, low, high)
    sketch = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)
    return sketch

input_dir = Path("/kaggle/input/datasets/badasstechie/celebahq-resized-256x256/celeba_hq_256")
images = sorted(input_dir.glob("*.jpg"))

train_images = images[:27000]
test_images  = images[27000:]

for split, split_images in [("train", train_images), ("test", test_images)]:
    dir_A = Path(f"/kaggle/working/pairs/{split}_A")
    dir_B = Path(f"/kaggle/working/pairs/{split}_B")
    dir_A.mkdir(parents=True, exist_ok=True)
    dir_B.mkdir(parents=True, exist_ok=True)

    for img_path in tqdm(split_images, desc=split):
        img = cv2.imread(str(img_path))
        if img is None:
            continue
        img = cv2.resize(img, (256, 256))
        sketch = canny_sketch(img)
        cv2.imwrite(str(dir_A / img_path.name), sketch)  # sketch → A
        cv2.imwrite(str(dir_B / img_path.name), img)     # photo  → B

print("Done.")
```
